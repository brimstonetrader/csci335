```python
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer

model = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M")
tokenizer = M2M100Tokenizer.from_pretrained("facebook/m2m100_418M")
```

```python
tokenizer.src_lang = "en"
encoded_zh = tokenizer("I am sad because I was just robbed.", return_tensors="pt")
generated_tokens = model.generate(**encoded_zh, forced_bos_token_id=tokenizer.get_lang_id("es"))
tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

```python
tokenizer.src_lang = "es"
encoded_zh = tokenizer("Estoy muy triste porque me robaron.", return_tensors="pt")
generated_tokens = model.generate(**encoded_zh, forced_bos_token_id=tokenizer.get_lang_id("en"))
tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

```python
def translations(nput):
    tokenizer.src_lang = "en"
    encoded = tokenizer(nput, return_tensors="pt")
    generated_tokens_ar = model.generate(**encoded, forced_bos_token_id=tokenizer.get_lang_id("ar"))
    generated_tokens_de = model.generate(**encoded, forced_bos_token_id=tokenizer.get_lang_id("de"))
    generated_tokens_es = model.generate(**encoded, forced_bos_token_id=tokenizer.get_lang_id("es"))
    generated_tokens_he = model.generate(**encoded, forced_bos_token_id=tokenizer.get_lang_id("he"))
    generated_tokens_hi = model.generate(**encoded, forced_bos_token_id=tokenizer.get_lang_id("hi"))
    generated_tokens_ja = model.generate(**encoded, forced_bos_token_id=tokenizer.get_lang_id("ja"))
    generated_tokens_pl = model.generate(**encoded, forced_bos_token_id=tokenizer.get_lang_id("pl"))
    generated_tokens_sw = model.generate(**encoded, forced_bos_token_id=tokenizer.get_lang_id("sw"))
    generated_tokens_vi = model.generate(**encoded, forced_bos_token_id=tokenizer.get_lang_id("vi"))
    generated_tokens_zh = model.generate(**encoded, forced_bos_token_id=tokenizer.get_lang_id("zh"))
    output = []
    for gts in [generated_tokens_ar, generated_tokens_de, generated_tokens_es, generated_tokens_he, generated_tokens_hi, generated_tokens_ja, generated_tokens_pl, generated_tokens_sw, generated_tokens_vi, generated_tokens_zh]:
        output += tokenizer.batch_decode(gts, skip_special_tokens=True)
    return output
translations("pig")
```

```python
def language_guessing_game():
    print("Do you want to play the language guessing game? Input y or n.")
    user_input = input()
    if user_input == "n":
        return
    else: 
        
        l = np.random.randint(0,99)
        l2 = np.random.randint(1,33)
        l3 = np.random.randint(1,33)
        l4 = np.random.randint(1,33)
        language = languages[l]
        
        ls = [languages[l], languages[(l + l2) % 100], languages[(l + l2 + l3) % 100], languages[(l + l2 + l3 + l4) % 100]]
        np.random.shuffle(ls)
        language_code = language_codes[l]
        
        sentence = random_sentences[np.random.randint(0, 50)]
        print(f"English sentence: {sentence}")
        tokenizer.src_lang = "en"
        encoded = tokenizer(sentence, return_tensors="pt")
        generated_tokens = model.generate(**encoded, forced_bos_token_id=tokenizer.get_lang_id(language_code))
        print(f"Translated sentence: {tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]}")
        print(f"Was that 1. {ls[0]}, 2. {ls[1]}, 3. {ls[2]}, 4. {ls[3]} ?")
        answer = input()
        if ls[int(answer) - 1] == language:
            print("Correct!")
            language_guessing_game()
            return
        else: 
            print(f"Wrong! It was {language}")
            language_guessing_game()
            return
         
language_guessing_game()
```
